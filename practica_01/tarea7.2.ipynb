```python
import threading
import time

```

```python

boletos_disponibles = 1000
```

```python


# FUNCIÓN SIN LOCK 

def vender_boletos(cantidad):
    global boletos_disponibles
    temp = boletos_disponibles
    time.sleep(0.0001)  # simula latencia
    boletos_disponibles = temp - cantidad


# 1. EJECUCIÓN SIN LOCK

boletos_disponibles = 1000
hilos = []

for _ in range(100):  # 100 hilos
    hilo = threading.Thread(target=vender_boletos, args=(1,))
    hilos.append(hilo)
    hilo.start()

for hilo in hilos:
    hilo.join()

print("Boletos disponibles  SIN lock:", boletos_disponibles)

```

```python

# FUNCIÓN CON LOCK 

lock = threading.Lock()

def vender_boletos_lock(cantidad):
    global boletos_disponibles
    with lock:  # sección crítica protegida
        temp = boletos_disponibles
        time.sleep(0.0001)
        boletos_disponibles = temp - cantidad


# 2. EJECUCIÓN CON LOCK

boletos_disponibles = 1000
hilos = []

for _ in range(100):  # 100 hilos
    hilo = threading.Thread(target=vender_boletos_lock, args=(1,))
    hilos.append(hilo)
    hilo.start()

for hilo in hilos:
    hilo.join()

print("Boletos disponibles con Lock :", boletos_disponibles)
```
