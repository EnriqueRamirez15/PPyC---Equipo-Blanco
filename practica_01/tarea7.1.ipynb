```python
import time
import requests
import threading

```

```python
# Lista de ciudades
ciudades = [
    {"lat": 19.43, "lon": -99.13},  # CDMX
    {"lat": 40.71, "lon": -74.00},  # NY
    {"lat": 51.50, "lon": -0.12},   # Londres
    {"lat": 35.68, "lon": 139.69}   # Tokio
]

# Función para consultar el clima
def obtener_clima(lat, lon):
    base_url = "https://api.open-meteo.com/v1/forecast"
    params = f"?latitude={lat}&longitude={lon}&current_weather=true"
    url = base_url + params
    respuesta = requests.get(url)
    if respuesta.status_code == 200:
        return respuesta.json()['current_weather']
    return None
```

```python


# 1. EJECUCIÓN SECUENCIAL

inicio_secuencial = time.time()

for ciudad in ciudades:
    obtener_clima(ciudad['lat'], ciudad['lon'])

tiempo_secuencial = time.time() - inicio_secuencial
print(f"Tiempo secuencial: {tiempo_secuencial:.4f} segundos")


# 2. EJECUCIÓN CON HILOS

def tarea(ciudad):
    obtener_clima(ciudad['lat'], ciudad['lon'])

inicio_hilos = time.time()

hilos = []

for ciudad in ciudades:
    hilo = threading.Thread(target=tarea, args=(ciudad,))
    hilos.append(hilo)
    hilo.start()

# Esperar a que todos los hilos terminen
for hilo in hilos:
    hilo.join()

tiempo_hilos = time.time() - inicio_hilos
print(f"Tiempo con hilos: {tiempo_hilos:.4f} segundos")


# 3. COMPARACIÓN

print("\nComparación de resultados:")
if tiempo_hilos < tiempo_secuencial:
    print("La versión con hilos es más rápida.")
else:
    print("La versión secuencial fue más rápida (puede variar por la red).")
```
